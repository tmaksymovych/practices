У Pandas нам переважно цікаві дві структури: `Series` та `DataFrame`.

Почнемо знайомство з `Series`. З лекцій ви вже знаєте, що `Series` у чомусь схожий на комбінацію словника (`dict`) та списку (`list`).

Для початку створіть `Series` з іменем `s`, де індексами будуть літери "a", "b", "c", "d", а відповідними значеннями - 1, 7, 2, 10. Можете використати будь-який зручний вам спосіб.

```python
s = pd.Series([1, 7, 2, 10], index=['a', 'b', 'c', 'd'])
s
```

Крім того, нам знадобиться ще один `Series` на імʼя `t`, але вже з наступним мапінгом: a->3, c -> 10, x -> 3.2, d -> -1. Заодно подивіться, як змінився тип даних

```python
t = pd.Series([3, 10, 3.2, -1], index=['a', 'c', 'x', 'd'])
t
```

Сподіваюсь, усе вийшло. `Series` зручні тим, що ними можна оперувати, як вже звичними нам `numpy.ndarray`, але елементи будуть "метчитися" на основі їх індексів, а не просто місця у списку

Спробуйте додати `s` та `t`, збережіть результат у змінну `s_plus_t` і погляньте на результат.

```python
s_plus_t = s + t
s_plus_t
```

Бачимо, що у деяких елементів результат став `NaN`, тобто "Not a Number". Тут у нагоді стане функція pd.isna (саме `isna`, не `isnan`). Погляньте, що повертає ця функція при застосуванні її до `s + t`:

```python
pd.isna(s_plus_t)
```

Можливо, тепер здогадаєтесь, як відфільтрувати непотрібні записи? Підказка: працює усе дуже схожим чином на `numpy.ndarray` :) 

```python
s_plus_t[pd.isna(s_plus_t)]
```

Ще однією приємною новиною є те, що функції `numpy` цілком сумісні з Series. Порахуйте синус від елементів `s` і переконайтесь у цьому! 

```python
np.sin(s)
```

Тепер спробуйте виділити елементи від другого до передостаннього зі списку `s`. Тут все відбувається так само схожим чином з `numpy`.

```python
s[1:-1]
```
